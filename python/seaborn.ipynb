```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Seaborn best practices

## Create dummy data

```python
data = pd.read_csv(
    filepath_or_buffer="../_data/concurrent_bike_rentals_dataset.csv",
    parse_dates=["date"]
)
```

## Set style

```python
sns.set_theme()
sns.set(rc = {'figure.figsize':(10,6)})
```

```python
sns.boxplot(x="month", y="concurrent_rentals", data=data, color="lightgrey", width=0.5)
sns.stripplot(x="month", y="concurrent_rentals", data=data, alpha=0.5, hue="weekday")
```

```python
sns.ecdfplot(data, x="concurrent_rentals", hue="weekday")
```

```python
g = sns.FacetGrid(data=data, col="month")
g.map_dataframe(sns.ecdfplot, data=data, x="concurrent_rentals", hue="weekday")
```
